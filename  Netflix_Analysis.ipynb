```python
import pandas as pd
#Load the dataset 
df = pd.read_csv('netflix_titles.csv')

#Show fisrt 5 rows
df.head()
```

```python
df
```

```python
# Clean and understand the data 

# structure of data 
df.info()
```

```python
# missing values

df.isnull().sum()
```

```python
# unique content type 

df['type'].value_counts()
```

```python
# how many rows had invalid dates
df['date_added'].isnull().sum()
```

```python
# How many titles were released each year

import matplotlib.pyplot as plt 

df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_added'] = df['date_added'].dt.year

df['year_added'].value_counts().sort_index().plot(kind='bar' , figsize=(12,6))
plt.title('Number of Netflix titles added per year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

```python
# What are the most common genres

df['listed_in'].value_counts().head(10)
```

```python
# Visulize with Seaborn
import seaborn as sns

# Movies vs TV Shows 

sns.countplot(x='type',data=df)
plt.title('Movie VS TV Show')
plt.show()
```

```python
df.to_csv('cleaned_netflix_data.csv', index=False)
```

```python
# What are the most common content types

import seaborn as sns 

sns.countplot(data=df,x='type',hue='type',palette='Set2')
plt.title('Distribution of movies and tv shows')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
# Which countries have the most content

top_countries = df['country'].value_counts().head(10)

top_countries.plot(kind='barh',color='skyblue')
plt.title('Top 10 countries by netflix content')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.gca().invert_yaxis()
plt.show()
```

```python
# What are the most common ratings

df['rating'].value_counts().plot(kind='bar',color='coral',figsize=(10,5))
plt.title('Distribution of content ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# check which day/month Netflix adds most titles

df['month_added'] = df['date_added'].dt.month_name()
df['month_added'].value_counts().sort_values().plot(kind='barh', color='green')
plt.title('Titles Added by Month')
plt.xlabel('Count')
plt.ylabel('Month')
plt.show()
```
